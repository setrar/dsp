```julia
using Plots

# Define the parameters
N = 100
 = 0.2  # Adjust this value to see different results
n = 0:N-1

# Coefficients
a = 1 / (2im)
a = -1 / (2im)

# Signal
x = a * exp.(im *  * n) + a * exp.(-im *  * n)

# Plot the signal as picks (scatter plot)
scatter(n, real(x), line=:stem, marker=:dot, markersize=2, label="Real part", xlabel="n", ylabel="x[n]", legend=:top)
```

```julia
using Plots

# Define the parameters
N = 100
 = 0.2  # Adjust this value to see different results
n = -N/2:N-1

# Coefficients
a = 1 / (2im)
a = -1 / (2im)

# Signal
x = a * exp.(im *  * n) + a * exp.(-im *  * n)

# Plot the signal and coefficients on the same graph
scatter(n, real(x), line=:stem, marker=:dot, markersize=2, label="Signal Real part", xlabel="n", ylabel="x[n]", legend=:top)
scatter!([1, 2], [real(a), real(a)], marker=:dot, markersize=4, label="Coefficients Real part")
scatter!([1, 2], [imag(a), imag(a)], marker=:dot, markersize=4, label="Coefficients Imaginary part")

# Center the x-axis
# xticks!([(1 + 2)/2], [""])
```

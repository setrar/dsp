```julia
using Plots

# Define the function for x(t)
function x(t, a)
    return exp(-a * t) * (t >= 0 ? 1 : 0)
end

# Set the value of 'a' (choose any positive value)
a_value = 1

# Generate time values
t_values = LinRange(0, 5, 1000)

# Calculate the corresponding x(t) values
x_values = x.(t_values, a_value)

# Plot x(t)
plot(t_values, x_values, xlabel="t", ylabel="x(t)", label="x(t)", legend=:topleft)
title!("Time-domain signal x(t) = e^{-at} * u(t)")
```

```julia
using Plots
using LaTeXStrings

# Function to calculate the Fourier transform
X(ω, a) = 1 / (a + ω*im)

# Set the value of 'a'
aₓ = 1

# Frequency range
wₓ = LinRange(-10, 10, 1000)

# Calculate the Fourier transform for the given 'a'
Xᵥᵥ = X.(wₓ, aₓ)

# Plot the magnitude
plot(wₓ, abs.(Xᵥᵥ), xlabel="Frequency (ω)", ylabel="|X(ω)|", label="Magnitude", legend=:topleft)
title!("Magnitude of Fourier Transform")
```

```julia
# Plot the phase
plot(wₓ, angle.(Xᵥᵥ), xlabel="Frequency (ω)", ylabel="Phase of X(ω)", label="Phase", legend=:topleft)
title!("Phase of Fourier Transform")
```

```julia
using Plots

# Define the function for x(t)
function x(t, a)
    return exp(-a * abs(t))
end

# Set the value of 'a' (choose any positive value)
a_value = 1

# Generate time values
t_values = LinRange(-5, 5, 1000)

# Calculate the corresponding x(t) values
x_values = x.(t_values, a_value)

# Plot x(t)
plot(t_values, x_values, xlabel="t", ylabel="x(t)", label="x(t)", legend=:topleft)
title!("Time-domain signal x(t) = e^{-a|t|}")
```

```julia
using Plots

# Define the function for X(ω)
function X(ω, a)
    return 2a / (a^2 + ω^2)
end

# Set the value of 'a' (choose any positive value)
a_value = 1

# Generate frequency values
ω_values = LinRange(-5, 5, 1000)

# Calculate the corresponding X(ω) values
X_values = X.(ω_values, a_value)

# Plot the magnitude of X(ω)
plot(ω_values, abs.(X_values), xlabel="ω", ylabel="|X(ω)|", label="Magnitude of X(ω)", legend=:topleft)
title!("Magnitude of Fourier Transform X(ω)")
```

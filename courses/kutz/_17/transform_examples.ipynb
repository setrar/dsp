```julia
using Plots

# Define the function for x(t)
function x(t, a)
    return exp(-a * t) * (t >= 0 ? 1 : 0)
end

# Set the value of 'a' (choose any positive value)
a_value = 1

# Generate time values
t_values = LinRange(0, 5, 1000)

# Calculate the corresponding x(t) values
x_values = x.(t_values, a_value)

# Plot x(t)
plot(t_values, x_values, xlabel="t", ylabel="x(t)", label="x(t)", legend=:topleft)
title!("Time-domain signal x(t) = e^{-at} * u(t)")
```

```julia
using Plots
using LaTeXStrings

# Function to calculate the Fourier transform
X(, a) = 1 / (a + *im)

# Set the value of 'a'
a = 1

# Frequency range
w = LinRange(-10, 10, 1000)

# Calculate the Fourier transform for the given 'a'
X = X.(w, a)

# Plot the magnitude
plot(w, abs.(X), xlabel="Frequency ()", ylabel="|X()|", label="Magnitude", legend=:topleft)
title!("Magnitude of Fourier Transform")
```

```julia
# Plot the phase
plot(w, angle.(X), xlabel="Frequency ()", ylabel="Phase of X()", label="Phase", legend=:topleft)
title!("Phase of Fourier Transform")
```

```julia
using Plots

# Define the function for x(t)
function x(t, a)
    return exp(-a * abs(t))
end

# Set the value of 'a' (choose any positive value)
a_value = 1

# Generate time values
t_values = LinRange(-5, 5, 1000)

# Calculate the corresponding x(t) values
x_values = x.(t_values, a_value)

# Plot x(t)
plot(t_values, x_values, xlabel="t", ylabel="x(t)", label="x(t)", legend=:topleft)
title!("Time-domain signal x(t) = e^{-a|t|}")
```

```julia
using Plots

# Define the function for X()
function X(, a)
    return 2a / (a^2 + ^2)
end

# Set the value of 'a' (choose any positive value)
a_value = 1

# Generate frequency values
_values = LinRange(-5, 5, 1000)

# Calculate the corresponding X() values
X_values = X.(_values, a_value)

# Plot the magnitude of X()
plot(_values, abs.(X_values), xlabel="", ylabel="|X()|", label="Magnitude of X()", legend=:topleft)
title!("Magnitude of Fourier Transform X()")
```
